```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
shitomasi_df = pd.read_csv('resultCSV/Detector_Descriptor_TTC_Camera_Perf_v2.csv')
shitomasi_df
```

```python
# Plot camera based TTC times for each detector
shitomasi_brisk_ttc_times = shitomasi_df.loc[0,:]
shitomasi_brief_ttc_times = shitomasi_df.loc[1,:]
shitomasi_orb_ttc_times = shitomasi_df.loc[2,:]
shitomasi_freak_ttc_times = shitomasi_df.loc[3,:]
shitomasi_sift_ttc_times = shitomasi_df.loc[4,:]

plt.plot(shitomasi_brisk_ttc_times[1:shitomasi_brisk_ttc_times.size])
plt.plot(shitomasi_brief_ttc_times[1:shitomasi_brief_ttc_times.size])
plt.plot(shitomasi_orb_ttc_times[1:shitomasi_orb_ttc_times.size])
plt.plot(shitomasi_freak_ttc_times[1:shitomasi_freak_ttc_times.size])
plt.plot(shitomasi_sift_ttc_times[1:shitomasi_sift_ttc_times.size])

plt.legend([shitomasi_df.loc[0][0], shitomasi_df.loc[1][0], shitomasi_df.loc[2][0], shitomasi_df.loc[3][0], shitomasi_df.loc[4][0]])
plt.xlabel('Frame #')
plt.ylabel('TTC (in seconds)')
plt.ylim([5,20])
plt.grid('on')
plt.title('Camera-based TTC - Shi-Tomasi Detector')
```
